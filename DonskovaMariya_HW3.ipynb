```python
import math
import numpy as np

class Triangle:
    EPS = 1E-15
    
    def __init__(self, l1, l2, l3):
        self.l1 = l1
        self.l2 = l2
        self.l3 = l3
        if (l1 + l2 <= l3) or (l1 + l3 <= l2) or (l2 + l3 <= l1):
            raise Exception()
        
    def perimeter(self):
        return self.l1 + self.l2 + self.l3
    
    def area(self):
        p = self.perimeter() / 2
        return math.sqrt(p * (p-self.l1) * (p-self.l2) * (p-self.l3))
    
    def isosceles(self):
        return (math.isclose(self.l1, self.l2, rel_tol=self.EPS) or
            math.isclose(self.l2, self.l3, rel_tol=self.EPS) or
            math.isclose(self.l1, self.l3, rel_tol=self.EPS))

    def equilateral(self):
        return (math.isclose(self.l1, self.l2, rel_tol=self.EPS) and
            math.isclose(self.l2, self.l3, rel_tol=self.EPS))
    
    def __eq__(self, other):
        tr_1 = sorted([self.l1, self.l2, self.l3])
        tr_2 = sorted([other.l1, other.l2, other.l3])
        return (math.isclose(tr_1[0], tr_2[0], rel_tol=self.EPS) and
            math.isclose(tr_1[1], tr_2[1], rel_tol=self.EPS) and
            math.isclose(tr_1[2], tr_2[2], rel_tol=self.EPS))
    
```

```python
import numpy as np
import math

class Item:
    def __init__(self, name, price, category):
        self.name = name
        self.price = price
        self.category = category

class Shop:
    def __init__(self):
        self.items = []
        
    def add(self, item):
        self.items.append(item)
        
    def count_category(self, category):
        n = 0
        for i in self.items:
            if i.category == category:
                n = n + 1
        return n
    def find_items(self, price_bounds, categories):
        lst_sort = []
        if price_bounds[0] == None:
            price_bounds[0] = -math.inf
        if price_bounds[1] == None:
            price_bounds[1] = math.inf
        for i in self.items:
            if categories == None:
                if (i.price >= price_bounds[0]) and (i.price <= price_bounds[1]):
                    lst_sort.append(i)
            else:
                if i.category in categories:
                    if (i.price >= price_bounds[0]) and (i.price <= price_bounds[1]):
                        lst_sort.append(i)
        return lst_sort
    
    def get_full_price(self, categories):
        full_price = 0
        if categories != None:
            for i in self.items:
                if i.category in categories:
                    full_price = full_price + i.price
        else:
            for i in self.items:
                full_price = full_price + i.price
        return full_price
    
    def is_available(self, item):
        return item in self.items
        
    def __contains__(self, item):
        return item in self.items
```
